# Paso 3. Construir un modelo para predecir la cancelación de usuarios
Crea un modelo de clasificación binaria para clientes donde la característica objetivo es la marcha del usuario o la usuaria el mes siguiente

### - Divide los datos en conjuntos de entrenamiento y validación utilizando la función train_test_split().

```python
# Dividiendo los datos en características (matriz x) y la variable objetivo (y)
X= data.drop('churn', axis=1)
y= data['churn']
```

```python
# Dividiendo los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

### - Entrena el modelo en el set de entrenamiento con dos métodos:
    - regresión logística;
    - bosque aleatorio.

```python
# Entrenando StandardScaler en el conjunto de entrenamiento:
scaler= StandardScaler()
scaler.fit(X_train)
# Transformando conjuntos de entrenamiento y de prueba:
X_train_st= scaler.transform(X_train)
X_test_st= scaler.transform(X_test)
```

#### - Regresión Logística

```python
# Definiendo el algoritmo del modelo:
lr_model = LogisticRegression(random_state=0)#, max_iter=10000)  
# Entrenenado el modelo:
lr_model.fit(X_train_st, y_train)
# Utilizando el modelo para hacer predicciones:
lr_predictions= lr_model.predict(X_test_st)
lr_probabilities= lr_model.predict_proba(X_test_st)[:,1]
```

#### - Bosque Aleatorio

```python
# Definiendo el algoritmo del modelo:
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)  
# Entrenenado el modelo:
rf_model.fit(X_train_st, y_train)
# Utilizando el modelo para hacer predicciones:
rf_predictions= rf_model.predict(X_test_st)
rf_probabilities= rf_model.predict_proba(X_test_st)[:,1]
# Estimando R cuadrado utilizando los datos de prueba:
rf_r2= r2_score(y_test, rf_predictions)
print('Valor R cuadrado para Random Forest: ', rf_r2)
```

### - Evalúa la exactitud, precisión y recall para ambos modelos utilizando los datos de validación. Utilízalos para comparar los modelos. ¿Qué modelo dio mejores resultados?

```python
# Ddefiniendo la función que podrá generar las métricas:
def print_all_metrics(y_true, y_pred, y_proba, title= 'Métricas de clasificación'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

##### Para Regresión Logística:

```python
# Mostrando todas las métricas.
print('Tamaño del dataset:', data.shape)
print_all_metrics(y_test, lr_predictions, lr_probabilities, title= 'Métricas de regresión logística')
```

##### Para Random Forest:

```python
# Mostrando todas las métricas.
print('Tamaño del dataset:', data.shape)
print_all_metrics(y_test, rf_predictions, rf_probabilities, title= 'Métricas de Random Forest')
```

## Conclusiones del Paso 3:

>El modelo de Regresión logística dió, aunque en una muy pequeña proporción, mejores resultados que el de Random Forest, Especificamente en precisión y recall, ya que en exactitud tienen el mismo nivel. Sin embargo, Como se ha mencionado, la diferencia es muy pequeña y se podría tomar en cuenta otros factores para una decisión más concluyente.
